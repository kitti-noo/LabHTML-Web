```
#if else
num = int(input('Enter your number?:'))
if num > 10:
  print('number > 10')
else:
  print('number <= 10')
  
```

```
#if elif else
num = int(input('Enter your number?:'))
if num > 10:
  print('number > 10')
elif num == 10:
  print('number = 10')
else:
  print('number < 10')
  
```
