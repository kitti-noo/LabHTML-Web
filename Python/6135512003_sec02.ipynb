```
#check3
n = int(input('Input number of Fibonacci number: '))
for i in range (0,n+1):
  def fib(i):
    if i== 0 :
      return 0
    elif i == 1:
      return 1
    else:
      return fib(i-1) + fib(i-2)
  if i == 10:
    print(fib(i))
  else:
    print(fib(i) , end=', ' )


  
 
```
